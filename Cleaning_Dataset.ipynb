```python
# Definisikan Token Gejala

import regex as re

sentence = []
token_gejala = []
word_exclude = [
    'informasi', 'obat', 'ini', 'hanya', 'untuk', 'kalangan', 'medis', 'meredakan', 'gejala', 'yang', 'berhubungan', 'dengan', 'misalnya', 'alergi', 'dan', 'atau',
    'mg', 'terapi', 'pada', 'pasien', 'tablet', 'digunakan', 'meringankan', 'ml', 'gejala-gejala', 'di', 'disertai', 'seperti', 'karena', 'sebagai', 'khususnya', 'mengobati', 'krn', 'serta', 'rasa', 'informasi', 'informasi'
]

for index, row in selected_data_frame.iterrows():

  all_text = row['indikasi-umum']

  # Memisahkan kata-kata dengan spasi sebagai pemisah
  words = all_text.split()


  # Menghilangkan tanda baca dari setiap kata, mengubah ke lowercase, mengabaikan kata yang mengandung angka, dan mengabaikan kata yang termasuk word_exclude
  words = [word.replace('&', ' ').strip(';/,.&()').lower() for word in words if word != '&' and not re.search(r'\d', word)]

  # Append data to token_gejala
  for word in words:
    if word not in token_gejala and word not in word_exclude:
      token_gejala.append(word)

# print(token_gejala)

print('Number of token_gejala : ', len(token_gejala))
for word in token_gejala[:20]:
  print(word)
```

```python
# Create a new column to include the token_gejala

import re
import pandas as pd

# Create a copy of the DataFrame
selected_data_frame_copy = selected_data_frame.copy()

# Create new columns for each gejala (symptom)
for gejala in token_gejala:
    pattern = r"\b" + re.escape(gejala) + r"\b"
    selected_data_frame_copy.loc[selected_data_frame_copy['indikasi-umum'].str.contains(pattern), gejala] = 1
    selected_data_frame_copy.loc[~selected_data_frame_copy['indikasi-umum'].str.contains(pattern), gejala] = 0

# Assign the modified DataFrame back to test_data
selected_data_frame = selected_data_frame_copy
```

```python
selected_data_frame.to_csv('output.csv', index=False)
```
