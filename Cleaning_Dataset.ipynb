```python
# Read data of word_excluded that will not be included as token_gejala
# Upload and read txt 
 
load_data()

with open("word_exclude_quote.txt" , "r") as file:
  txtFile = file.read()

  # Split the txtFile by commas
  tokens = txtFile.split(", ")

  # Remove leading and trailing single quotes from each token
  word_exclude = [token.strip("'") for token in tokens]

print(word_exclude[25:])
```

```python
# Definisikan Token Gejala

import regex as re

sentence = []
token_gejala = []

for index, row in selected_data_frame_2.iterrows():

  all_text = row['indikasi-umum']

  # Memisahkan kata-kata dengan spasi sebagai pemisah
  words = all_text.split()

  # Menghilangkan tanda baca dari setiap kata, mengubah ke lowercase, mengabaikan kata yang mengandung angka, dan mengabaikan kata yang termasuk word_exclude
  words = [(word.replace('/', ' ').replace('&', ' ')).strip('!@#$%/<>^;*:+-,.&()').lower() for word in words if word != '&' and not re.search(r'\d', word)]

  # Append data to token_gejala
  for word in words:
    split_word = word.split()
    for splitted_word in split_word:
      if splitted_word not in token_gejala and splitted_word not in str(word_exclude):
        token_gejala.append(splitted_word) 


print('Number of token_gejala : ', len(token_gejala))
# print('token_gejala preview : ')
for word in token_gejala[:25]:
  print(word)

with open('token_gejala.txt', 'w') as output_file:
  for word in token_gejala:
    output_file.write(word + '\n')
```

```python
# Create a new column to include the token_gejala

# Create a copy of the DataFrame
selected_data_frame_copy = selected_data_frame_2.copy()

# Create new columns for each gejala (symptom)
for gejala in token_gejala:
    pattern = r"\b" + re.escape(gejala) + r"\b"
    selected_data_frame_copy.loc[selected_data_frame_copy['indikasi-umum'].str.contains(pattern), gejala] = 1
    selected_data_frame_copy.loc[~selected_data_frame_copy['indikasi-umum'].str.contains(pattern), gejala] = 0

# Assign the modified DataFrame back to df
selected_data_frame_2 = selected_data_frame_copy
```

```python
selected_data_frame_copy.to_csv('drugs_cleaned_dataset.csv', index=False)
```
